## Candidate Key Assessment

Identify potential primary keys and business identifiers.

```python
for name, df in datasets.items():

    print("="*70)
    print(name)

    for column in df.columns:

        unique = df[column].nunique()

        if unique == len(df):

            print(f"{column} → Candidate Primary Key")
```

> **Note**
>
> The candidate key assessment performed in this notebook is based solely on **uniqueness** and should not be interpreted as confirmation of actual database primary keys.
>
> - In the **Orders** dataset, `customer_id` is unique but functions as a **foreign key** referencing the Customers dataset rather than a primary key. The actual primary identifier is `order_id`.
> - In the **Product Category Translation** dataset, both `product_category_name` and `product_category_name_english` contain unique values. They are therefore **candidate keys**, but neither should be assumed to be the table's primary key unless explicitly defined by the database schema.
>
> The final classification of primary keys, foreign keys, and table relationships will be confirmed during the **Relationship Analysis** phase.

# 10. Initial Observations

## Customers

- Contains one record per customer with geographic information at ZIP code, city, and state levels.
- `customer_id` is unique for every record and is a candidate primary key.
- `customer_unique_id` has fewer unique values than `customer_id`, indicating that a customer may place multiple orders.
- No missing values were identified.
- Dataset appears suitable as a customer dimension table.

---

## Geolocation

- Largest dataset in the project with over one million records.
- Provides geographic coordinates mapped to ZIP code prefixes.
- No missing values were identified.
- The dataset contains a large number of duplicate records, which may require deduplication during data preparation.
- Geographic information may be used for regional analysis and mapping.

---

## Order Items

- Contains one record per product purchased within an order.
- `order_id` is not unique, confirming that a single order can contain multiple products.
- Includes product, seller, pricing, freight, and shipping information.
- No missing values were identified.
- `shipping_limit_date` is currently stored as an object and will require datetime conversion.
- Dataset represents the lowest transactional grain within the sales process.

---

## Payments

- Stores payment information associated with customer orders.
- Multiple payment records exist for some orders, indicating that an order can have more than one payment transaction.
- Credit card is the most frequently used payment method.
- No missing values were identified.
- Payment amount will require validation against order totals during relationship analysis.

---

## Reviews

- Contains customer satisfaction information through review scores and optional review comments.
- Review score is complete for all records.
- Review titles and review messages contain a large number of missing values, suggesting that written feedback is optional.
- Review-related date fields are stored as object data types.
- Dataset appears suitable for customer satisfaction analysis.

---

## Orders

- Appears to contain one record per order.
- `order_id` is unique and is the primary identifier.
- Contains the complete order lifecycle, including purchase, approval, shipping, delivery, and estimated delivery timestamps.
- Missing values exist in approval and delivery timestamp columns, likely corresponding to orders that were cancelled or not completed.
- All timestamp fields are currently stored as object data types and require datetime conversion.

---

## Products

- Contains product metadata, category information, and physical dimensions.
- `product_id` is unique for every record.
- Product category names are stored in Portuguese.
- Several product metadata columns contain missing values.
- Product dimensions and weight appear suitable for logistics-related analysis.

---

## Sellers

- Contains seller identifiers and geographic information.
- `seller_id` is unique for every record.
- No missing values were identified.
- Dataset appears suitable as a seller dimension table.

---

## Category Translation

- Provides mappings between Portuguese and English product category names.
- Both columns contain unique values.
- No missing values were identified.
- Required to standardize product categories for reporting and dashboard development.

---

# 11. Business Assumptions

The following assumptions are based on the initial dataset exploration and will be validated during relationship analysis.

- Orders is expected to be the central transactional dataset.
- Order Items represents the lowest level of sales transactions (one row per product purchased).
- Customers, Products, and Sellers function as dimension tables describing business entities.
- Payments contains financial transactions linked to customer orders.
- Reviews capture post-purchase customer feedback.
- Product Category Translation provides standardized English category names for reporting.
- Freight charges are stored separately from product prices.
- Geolocation provides reference data for geographic analysis rather than transactional information.

---

# 12. Potential Issues

The following observations require attention in later project phases.

- All date and timestamp columns are currently stored as object data types.
- Orders contains missing approval and delivery timestamps.
- Products contains missing category metadata and physical attribute values.
- Reviews contains a large number of missing review titles and review messages.
- Geolocation contains a significant number of duplicate records.
- Product categories are stored in Portuguese and require translation before reporting.
- Multiple payment records exist for some orders and require validation during relationship analysis.

No corrective actions are performed during this notebook.

```python
# ==========================================================
# Relationship Validation
# ==========================================================
# Objective:
# Validate that foreign keys in child tables exist in their
# corresponding parent tables and compare the number of
# unique keys between related datasets.
# ==========================================================

print("=" * 70)
print("RELATIONSHIP VALIDATION")
print("=" * 70)

# ----------------------------------------------------------
# Customers -> Orders
# ----------------------------------------------------------
print("\n[1] Customers -> Orders")
matched = orders['customer_id'].isin(customers['customer_id']).sum()

print(f"Matching customer_id values in Orders      : {matched:,}")
print(f"Unique customer_id in Orders               : {orders['customer_id'].nunique():,}")
print(f"Unique customer_id in Customers            : {customers['customer_id'].nunique():,}")

# ----------------------------------------------------------
# Orders -> Order Items
# ----------------------------------------------------------
print("\n[2] Orders -> Order Items")
matched = order_items['order_id'].isin(orders['order_id']).sum()

print(f"Matching order_id values in Order Items    : {matched:,}")
print(f"Unique order_id in Order Items             : {order_items['order_id'].nunique():,}")
print(f"Unique order_id in Orders                  : {orders['order_id'].nunique():,}")

# ----------------------------------------------------------
# Products -> Order Items
# ----------------------------------------------------------
print("\n[3] Products -> Order Items")
matched = order_items['product_id'].isin(products['product_id']).sum()

print(f"Matching product_id values in Order Items  : {matched:,}")
print(f"Unique product_id in Order Items           : {order_items['product_id'].nunique():,}")
print(f"Unique product_id in Products              : {products['product_id'].nunique():,}")

# ----------------------------------------------------------
# Sellers -> Order Items
# ----------------------------------------------------------
print("\n[4] Sellers -> Order Items")
matched = order_items['seller_id'].isin(sellers['seller_id']).sum()

print(f"Matching seller_id values in Order Items   : {matched:,}")
print(f"Unique seller_id in Order Items            : {order_items['seller_id'].nunique():,}")
print(f"Unique seller_id in Sellers                : {sellers['seller_id'].nunique():,}")

# ----------------------------------------------------------
# Orders -> Payments
# ----------------------------------------------------------
print("\n[5] Orders -> Payments")
matched = payments['order_id'].isin(orders['order_id']).sum()

print(f"Matching order_id values in Payments       : {matched:,}")
print(f"Unique order_id in Payments                : {payments['order_id'].nunique():,}")
print(f"Unique order_id in Orders                  : {orders['order_id'].nunique():,}")

# ----------------------------------------------------------
# Orders -> Reviews
# ----------------------------------------------------------
print("\n[6] Orders -> Reviews")
matched = reviews['order_id'].isin(orders['order_id']).sum()

print(f"Matching order_id values in Reviews        : {matched:,}")
print(f"Unique order_id in Reviews                 : {reviews['order_id'].nunique():,}")
print(f"Unique order_id in Orders                  : {orders['order_id'].nunique():,}")

# ----------------------------------------------------------
# Category Translation -> Products
# ----------------------------------------------------------
print("\n[7] Category Translation -> Products")
matched = products['product_category_name'].isin(
    category_translation['product_category_name']
).sum()

print(f"Matching product categories                : {matched:,}")
print(f"Unique categories in Products              : {products['product_category_name'].nunique():,}")
print(f"Unique categories in Translation Table     : {category_translation['product_category_name'].nunique():,}")

print("\n" + "=" * 70)
print("Relationship validation completed.")
print("=" * 70)
```

```python
# ==========================================================
# Data Integrity Validation
# ==========================================================
# Objective:
# Validate referential integrity by checking:
#   1. Orphan records (foreign keys with no matching parent)
#   2. Duplicate values in primary key columns
# ==========================================================

print("=" * 70)
print("DATA INTEGRITY VALIDATION")
print("=" * 70)

# ----------------------------------------------------------
# Orphan Record Validation
# ----------------------------------------------------------
print("\n[1] Orphan Record Validation")

print(
    f"Orders without matching Customers         : "
    f"{orders[~orders['customer_id'].isin(customers['customer_id'])].shape[0]:,}"
)

print(
    f"Order Items without matching Orders       : "
    f"{order_items[~order_items['order_id'].isin(orders['order_id'])].shape[0]:,}"
)

print(
    f"Payments without matching Orders          : "
    f"{payments[~payments['order_id'].isin(orders['order_id'])].shape[0]:,}"
)

print(
    f"Reviews without matching Orders           : "
    f"{reviews[~reviews['order_id'].isin(orders['order_id'])].shape[0]:,}"
)

# ----------------------------------------------------------
# Duplicate Primary Key Validation
# ----------------------------------------------------------
print("\n[2] Duplicate Join Key Validation")

print(
    f"Duplicate order_id in Orders              : "
    f"{orders['order_id'].duplicated().sum():,}"
)

print(
    f"Duplicate customer_id in Customers        : "
    f"{customers['customer_id'].duplicated().sum():,}"
)

print(
    f"Duplicate product_id in Products          : "
    f"{products['product_id'].duplicated().sum():,}"
)

print(
    f"Duplicate seller_id in Sellers            : "
    f"{sellers['seller_id'].duplicated().sum():,}"
)

print("\n" + "=" * 70)
print("Data integrity validation completed.")
print("=" * 70)
```
